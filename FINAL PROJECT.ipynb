```python
d = y_test - y_prediction
dataframe = pd.DataFrame(d)
dataframe = dataframe.reset_index(drop=True)
dataframe.rename(columns={'sum_of_first_8_columns': 'Difference between actual and predicted value'}, inplace=True)
dataframe.tail()
```

```python
import matplotlib.pyplot as plt
from scipy.stats import linregress
from scipy.stats import norm
plt.plot(dataframe.index, dataframe['Difference between actual and predicted value'])
plt.title('Deviation between actual and predicted')
plt.xlabel('№ of sample')
plt.ylabel('Difference in Watts')
plt.grid(True, linestyle='--', alpha=0.6)
plt.annotate('due to data outliers', xy=(95, -6900), xytext=(60, -5000),
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.5'))
```

```python
# Extract values from the DataFrame
values = dataframe['Difference between actual and predicted value']

# Calculate mean and standard deviation
mean = values.mean()
std_dev = values.std()

# Create a range of x-values
x = np.linspace(mean - 3 * std_dev, mean + 3 * std_dev, 1000)

# Calculate the probability density function (PDF) of the Gaussian distribution
pdf = norm.pdf(x, loc=mean, scale=std_dev)

# Create the plot
plt.plot(x, pdf, color='blue', label='Gaussian Distribution')

# Find the x-values for the central 80% of the data
lower_bound = norm.ppf(0.1, loc=mean, scale=std_dev)
upper_bound = norm.ppf(0.9, loc=mean, scale=std_dev)

# Fill the area representing 80% of the data
plt.fill_between(x, pdf, 0, where=(x >= lower_bound) & (x <= upper_bound), color='lightgray', alpha=0.5, label='80% of Data')

# Set axis labels and a title
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Gaussian Distribution')

# Show a grid
plt.grid(True, linestyle='--', alpha=0.6)
```

```python
# model assessment
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
mae = mean_absolute_error(y_test, y_prediction)
mse = mean_squared_error(y_test, y_prediction)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_prediction)
print("MAE:",mae)
print("MSE:",mse)
print("RMSE:",rmse)
print('R^2:',r2)
```
